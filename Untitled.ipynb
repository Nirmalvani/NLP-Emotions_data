### Cleaning text

```python
df_new
```

```python
def clean_text(text):
    new_text = text.lower() # converting text in lower
    stop = set(stopwords.words("english"))
    new_text = " ".join(word for word in new_text.split() if word.lower() not in stop) # Removing all the stopwords
    new_text = clean(new_text,no_urls=True,no_emoji=True,no_punct=True) # Removing all the punctuation
    lemma = WordNetLemmatizer()
    new_text = " ".join(lemma.lemmatize(word) for word in word_tokenize(new_text)) # Lemmatizing the words
    
    return new_text # Completely cleaned text
```

```python
df_new['clean_tweet'] = df_new['tweet'].apply(clean_text)
```

```python
tf1_train_ar = tf1_train.toarray()
tf2_test_ar = tf2_test.toarray()
```

```python
# Build a simple neural network model
ann = Sequential([
    Dense(128, activation='sigmoid', input_shape=(tf1_train_ar.shape[1],)),
    Dropout(0.1),
    Dense(64, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])

# Compile the model
ann.compile(optimizer=Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = ann.fit(tf1_train_ar, y_train, epochs=2, batch_size=32, validation_split=0.3)
```

```python
pred4 = ann.predict(tf2_test_ar)
```

```python
metric(y_test,y_pred_new)
```
